```python
import numpy as np
import operator
```

```python
#读取数据
f=open("./Iris_Data_Set.txt")
arraylines = f.readlines()#逐行读取数据，存储在一个list中
numberoflines = len(arraylines)
```

```python
#数据初始化
trainingData = np.zeros([int(numberoflines/5)*4,4])#新建矩阵，存储4个属性值
validatingData=np.zeros([int(numberoflines/5),4])
trainingClass = []#类型：numpy.ndarray-n维数组
validatingClass=[]
print(trainingData)
print(trainingClass)
print(validatingData)
print(validatingClass)
```

```python
trainingIndex = 0
validatingIndex=0
flag=0
for line in arraylines:
    line=line.strip('\n')#去除行末的换行符
    listFromLine = line.split(',')#type:list
    if flag%5!=0:#训练数据
        print(trainingIndex)
        trainingData[trainingIndex,:] = listFromLine[0:4]#第1到4个
        if listFromLine[-1]=="Iris-setosa":
            trainingClass.append(1)
        elif listFromLine[-1]=="Iris-versicolor":
            trainingClass.append(2)
        else:
            trainingClass.append(3)
        trainingIndex+=1
    else:
        print(validatingIndex)
        validatingData[validatingIndex,:]=listFromLine[0:4]
        if listFromLine[-1]=="Iris-setosa":
            validatingClass.append(1)
        elif listFromLine[-1]=="Iris-versicolor":
            validatingClass.append(2)
        else:
            validatingClass.append(3)
        validatingIndex+=1
    flag+=1
```

```python
print(trainingData)
print(trainingClass)
print(validatingData)
print(validatingClass)
```

```python
#选择合适的k值
k=4
correct=0
validatingIndex=0
for v in validatingData:
    dist=[]
    index=0
    for t in trainingData:
        distance=pow(((v-t)*(v-t)).sum(),0.5)#欧式距离
        dist.append([index,distance])
        index+=1
    dist.sort(key=lambda dist: dist[1]) #根据dist排序
    #选择dist最小k个
    dist=dist[0:k]
    #print(dist)
    classes=[0,0,0]
    for i in dist:
        #print(i[0])
        if trainingClass[i[0]]==1:
            classes[0]+=1
        elif trainingClass[i[0]]==2:
            classes[1]+=1
        else:
            classes[2]+=1
    #print(classes)
    #返回最可能的类别的索引
    predicted_value=classes.index(max(classes))+1
    actual_value=validatingClass[validatingIndex]
    if predicted_value==actual_value:
        correct+=1
    #else:
        #print(validatingIndex)
        #print(predicted_value)
        #print(actual_value)
    validatingIndex+=1
print("correct_rate=%f"%(correct/30))
```
